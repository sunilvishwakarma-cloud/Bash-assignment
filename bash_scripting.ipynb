1. Write a Bash script thatg
- Creates a directory named "backup" in the user's home directory^
- Copies all .txt files from the current directory into the "backup" directory^
- Appends the current date and time to the filenames of the copied files.

````python
...
#!/bin/bash

# Create the backup directory in the user's home directory
BACKUP_DIR="$HOME/backup"
mkdir -p "$BACKUP_DIR"

# Get the current date and time in a filename-friendly format
TIMESTAMP=$(date +"%Y%m%d_%H%M%S")

# Copy all .txt files from the current directory
for file in *.txt; do
    # Skip if no .txt files exist
    [ -e "$file" ] || continue

    # Extract filename without extension
    base_name="${file%.txt}"

    # Copy with timestamp appended to the filename
    cp "$file" "$BACKUP_DIR/${base_name}_${TIMESTAMP}.txt"
done

echo "Backup completed successfully."
...
````

 2. Create a script that
- Checks the system’s CPU and memory usage^
- Reports if the CPU usage is above 80% or if the available memory is below 20%^
- Logs the results to a file named system_health.log.

````python
...
```bash
#!/bin/bash

LOG_FILE="system_health.log"

# Get CPU usage (Linux)
CPU_USAGE=$(top -bn1 | grep "Cpu(s)" | awk '{print int(100 - $8)}')

# Get memory usage
MEM_TOTAL=$(free | awk '/Mem:/ {print $2}')
MEM_AVAILABLE=$(free | awk '/Mem:/ {print $7}')
MEM_AVAILABLE_PERCENT=$((MEM_AVAILABLE * 100 / MEM_TOTAL))

# Get current timestamp
TIMESTAMP=$(date +"%Y-%m-%d %H:%M:%S")

# Log header
echo "[$TIMESTAMP]" >> "$LOG_FILE"
echo "CPU Usage: ${CPU_USAGE}%" >> "$LOG_FILE"
echo "Available Memory: ${MEM_AVAILABLE_PERCENT}%" >> "$LOG_FILE"

# Check CPU threshold
if [ "$CPU_USAGE" -gt 80 ]; then
    echo "WARNING: CPU usage is above 80%!" >> "$LOG_FILE"
fi

# Check available memory threshold
if [ "$MEM_AVAILABLE_PERCENT" -lt 20 ]; then
    echo "WARNING: Available memory is below 20%!" >> "$LOG_FILE"
fi

echo "----------------------------------------" >> "$LOG_FILE"

echo "System health check completed. Results saved to $LOG_FILE."
```
...
````

3.Write a script that
- Reads a list of usernames from a file (e.g., user_list.txt)^
- Creates a new user for each username^
- Generates a random password for each user and saves the username and password to a file named
credentials.txt.

````python
...
```bash
#!/bin/bash

USER_FILE="user_list.txt"
CREDENTIALS_FILE="credentials.txt"

# Clear or create the credentials file
> "$CREDENTIALS_FILE"

# Read usernames line by line
while IFS= read -r username; do
    # Skip empty lines
    [ -z "$username" ] && continue

    # Create the user
    useradd -m "$username"

    # Generate a random 12-character password
    password=$(openssl rand -base64 12)

    # Set the user's password
    echo "${username}:${password}" | chpasswd

    # Save credentials
    echo "${username}:${password}" >> "$CREDENTIALS_FILE"

    echo "Created user: $username"
done < "$USER_FILE"

echo "User creation completed."
echo "Credentials saved in $CREDENTIALS_FILE"
```
...
````

````python
...
```bash
#!/bin/bash

# Prompt the user for the directory path
read -p "Enter the directory path to back up: " DIR_PATH

# Check if the directory exists
if [ ! -d "$DIR_PATH" ]; then
    echo "Error: Directory does not exist."
    exit 1
fi

# Get the current date
CURRENT_DATE=$(date +"%Y-%m-%d")

# Create the backup filename
BACKUP_FILE="backup_${CURRENT_DATE}.tar.gz"

# Compress the directory
tar -czf "$BACKUP_FILE" -C "$(dirname "$DIR_PATH")" "$(basename "$DIR_PATH")"

echo "Backup completed successfully."
echo "Compressed file created: $BACKUP_FILE"
```
...
````

5.Create a Bash script that
- Implements a simple command-line to-do list^
- Allows the user to add tasks, view tasks, and remove tasks^
- Saves the tasks to a file (e.g., todo.txt).

````python
...
```bash
#!/bin/bash

TODO_FILE="todo.txt"

# Create the file if it doesn't exist
touch "$TODO_FILE"

while true; do
    echo "=========================="
    echo "      TO-DO LIST MENU"
    echo "=========================="
    echo "1. View Tasks"
    echo "2. Add Task"
    echo "3. Remove Task"
    echo "4. Exit"
    echo "=========================="

    read -p "Choose an option (1-4): " choice

    case $choice in
        1)
            if [ ! -s "$TODO_FILE" ]; then
                echo "No tasks found."
            else
                echo "Current Tasks:"
                nl -w2 -s'. ' "$TODO_FILE"
            fi
            ;;
        2)
            read -p "Enter the task: " task
            echo "$task" >> "$TODO_FILE"
            echo "Task added successfully."
            ;;
        3)
            if [ ! -s "$TODO_FILE" ]; then
                echo "No tasks to remove."
            else
                nl -w2 -s'. ' "$TODO_FILE"
                read -p "Enter the task number to remove: " num
                sed -i "${num}d" "$TODO_FILE"
                echo "Task removed successfully."
            fi
            ;;
        4)
            echo "Goodbye!"
            exit 0
            ;;
        *)
            echo "Invalid option. Please choose between 1 and 4."
            ;;
    esac

    echo
done
```
...
````

6.Write a script that
- Reads a list of software package names from a file (e.g., packages.txt)^
- Installs each package using the appropriate package manager (apt, yum, etc.)^
- Logs the installation status of each package.

````python
...
```bash id="itdwh8"
#!/bin/bash

PACKAGE_FILE="packages.txt"
LOG_FILE="installation.log"

# Check if the package list exists
if [ ! -f "$PACKAGE_FILE" ]; then
    echo "Error: $PACKAGE_FILE not found."
    exit 1
fi

# Detect package manager
if command -v apt >/dev/null 2>&1; then
    PKG_MANAGER="apt"
    INSTALL_CMD="apt install -y"
elif command -v dnf >/dev/null 2>&1; then
    PKG_MANAGER="dnf"
    INSTALL_CMD="dnf install -y"
elif command -v yum >/dev/null 2>&1; then
    PKG_MANAGER="yum"
    INSTALL_CMD="yum install -y"
else
    echo "No supported package manager found."
    exit 1
fi

echo "Using package manager: $PKG_MANAGER" > "$LOG_FILE"
echo "Installation started at $(date)" >> "$LOG_FILE"
echo "----------------------------------------" >> "$LOG_FILE"

# Read and install packages
while IFS= read -r package; do
    # Skip empty lines
    [ -z "$package" ] && continue

    echo "Installing $package..."

    if $INSTALL_CMD "$package" >/dev/null 2>&1; then
        echo "$(date): $package - SUCCESS" >> "$LOG_FILE"
    else
        echo "$(date): $package - FAILED" >> "$LOG_FILE"
    fi
done < "$PACKAGE_FILE"

echo "----------------------------------------" >> "$LOG_FILE"
echo "Installation process completed at $(date)" >> "$LOG_FILE"

echo "Done. Check $LOG_FILE for installation results."
```
...
````

- Finds and displays the longest word in the file.

````python
```bash
#!/bin/bash

# Prompt for the input file
read -p "Enter the text file name: " FILE

# Check if the file exists
if [ ! -f "$FILE" ]; then
    echo "Error: File '$FILE' not found."
    exit 1
fi

# Count lines, words, and characters
LINES=$(wc -l < "$FILE")
WORDS=$(wc -w < "$FILE")
CHARS=$(wc -m < "$FILE")

# Find the longest word
LONGEST_WORD=$(tr -c '[:alnum:]' '\n' < "$FILE" | awk '
{
    if (length($0) > max) {
        max = length($0)
        longest = $0
    }
}
END {
    print longest
}')

# Display results
echo "===== File Statistics ====="
echo "File       : $FILE"
echo "Lines      : $LINES"
echo "Words      : $WORDS"
echo "Characters : $CHARS"
echo "Longest Word: $LONGEST_WORD"
```
````
